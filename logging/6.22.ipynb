```python
salpha18=camcontrol.DataRun('data/6.21.skalpha18', gain = '0x3f',
        htime = '2h', run = False, block = True, window_min = 30, window_max = 70, photon_value=62.)
```

### Take another dark run:

```python
dark2=camcontrol.DataRun('data/6.22.dark2', gain = '0x3f',
        htime = '10m', run = False, block = True, window_min = 30, window_max = 70, photon_value=62.)
```

```python
def cleanplot(datarun, darkrun, **kwargs):
    clean = datarun.get_frame().remove_hot(darkrun)
    clean.plot_lineout(**kwargs)

f =  salpha18.get_frame()
fclean = f.remove_hot(darkrun = dark2)
```

### Load the old one-hour exposure that was taken without the slit in front of the analyzer:

```python
salpha_full_analyzer = camcontrol.DataRun('data/6.17.skalpha11', gain = '0x3f',
        htime = '1h', run = False, block = True, window_min = 30, window_max = 70, photon_value=62.)
```

### Plot the old (un-slitted) and new (slitted) data without and with hot pixel removal. 

The new peak is marginally narrower, and--more strangely--seems to have a different shape.

```python
_ = salpha18.plot_lineout(xmax = 1000, rebin = 4, show = False, peaknormalize = True)
_ = salpha_full_analyzer.plot_lineout(xmax = 1000, rebin = 4, peaknormalize = True)

cleanplot(salpha18, dark2, xmax = 1000, rebin = 4, show = False, peaknormalize = True)
cleanplot(salpha_full_analyzer, dark2, xmax = 1000, rebin = 4, peaknormalize = True)
_ = plot_width_progression([salpha_full_analyzer, salpha18], bragg_degrees = 59., energy = 2308.)
```
