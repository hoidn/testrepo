```python
# Reprocess 'long4' data from last time
prefixes = ['data/4.29.real%d' % i for i in range(36, 46)]
rset = camcontrol.RunSet(prefixes = prefixes)
reduced_frame = rset.filter_reduce_frames(threshold_min = 40, threshold_max = 100)
```

```python
reduced_frame.show(vmin = 0, vmax = 70, origin = 'nearest')
plt.plot(reduced_frame.lineout(smooth = 2))
plt.show()
```
